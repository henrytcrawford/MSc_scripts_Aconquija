```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pysptools
import pysptools.spectro as spectro
import pandas as pd
```

```python
df = pd.read_csv(r'c:\Users\htccr\Documents\Aconquija\Terraspec\spec_libs\prisma.csv')

df = df[0:230] # for PRISMA!
nm = np.round(df["nm"],1).tolist()
df = df.drop(['nm'], axis=1)
names = df.columns.tolist()

df = df - df.min() # for PRISMA!

len(names)
```

```python
# remove continum for each spectral profile
df_list = df.T.values.tolist()
n = len(names)
df_crs = []

for i in range(n):
    crs = spectro.convex_hull_removal(df_list[i], nm)
    df_crs.append(crs)

y=np.array(df_crs, dtype = list)

# store values
crs = []
cx = []
cy = []

for i in range(n):
    crs_vals = y[i][0]
    cx_vals = y[i][1]
    cy_vals = y[i][2]
    
    crs.append(crs_vals)
    cx.append(cx_vals)
    cy.append(cy_vals)
```

```python
# plot continums
plt.figure(figsize = (10,6))
for i in range(n):
    plt.plot(nm, df.T.iloc[i], label = names[i])
    plt.title('Continuim')
    plt.legend()
```

```python
# plot continums
plt.figure(figsize = (10,6))
for i in range(n):
    plt.plot(cx[i],cy[i], label = names[i])
    plt.title('Continuim')
    plt.legend()
```

```python
# plot continum removed spectra
plt.figure(figsize = (10,6))
for i in range(n):
    plt.plot(nm,crs[i], label = names[i])
    plt.title('Continuim Removed')
    #plt.legend()
    plt.xlim()
```

```python
# convert crs data to pandas, save as .csv
crs_df = pd.DataFrame(crs)
crs_df = crs_df.T
crs_df.columns = [names]

#crs_df.to_csv('terraspec_crs.csv', index=False, header = True) # save to files
```

```python
# convert contium values to same shape as nm i.e. interpoalte new y values at higher resolution.
cy_int = []
for i in range(n):
    interpolated_vals = np.interp(nm,cx[i],cy[i])
    cy_int.append(interpolated_vals)
```

```python
# convert contium data to pandas, save as .csv
c_df = pd.DataFrame(cy_int)
c_df = c_df.T
c_df.columns = [names]

#c_df.to_csv('terrspec_continuum.csv', index=False, header = True) # save to files
```
